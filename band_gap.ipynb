* now we will convert polymer formulas into arrays.

```python
polymer_ = [np.array(AllChem.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(i), 2)) for i in df.polymer]
```

```python
df = df.drop('polymer',axis=1)
```

```python
df.columns
```

```python
for column in df.columns:
    if df[column].dtype in [int,float]: 
        print(df[column].min(),(df[column].max()))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
numeric_columns = df.select_dtypes(include=['int64','float64']).columns
```

```python
scaled_data = scaler.fit_transform(df[numeric_columns])
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=numeric_columns)
```

```python
scaled_df = pd.concat([scaled_df, df.drop(columns=numeric_columns)], axis=1)
```

```python
scaled_df.head(5)
```

```python
polymer_column = scaled_df['polymer1']
flattened_arrays = pd.DataFrame(polymer_column.tolist(), columns=[f'element_{i}' for i in range(len(polymer_column[0]))])
processed_df = pd.concat([scaled_df.drop(columns=['polymer1']), flattened_arrays], axis=1)
processed_df
```

```python
X = processed_df.drop('band_gap',axis=1)
X
```

```python
y = processed_df['band_gap']
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

model_selection = {
    "LinearRegression" : {
        "model" : LinearRegression(),
        "params" : {}
    },
    "Ridge" :{
        "model" : Ridge(),
        "params" : {'alpha':[1,2,3,4,5,6,7,8,9]}
    },
    "Lasso" : {
        "model": Lasso(),
        "params":{'alpha':[1,2,3,4,5,6,7,8,9]}
    },
    "ElasticNet" : {
        "model": ElasticNet(),
        "params":{'alpha':[1,2,3,4,5,6,7,8,9],
                 'l1_ratio':[0.5,1,1.5,2,2.5,3]}
    },
    "DecisionTreeRegressor" : {
        "model": DecisionTreeRegressor(),
        "params":{"criterion": ["squared_error"],
                  'splitter':['random']}
    },
    "RandomForestRegressor" : {
        "model": RandomForestRegressor(),
        "params":{'n_estimators':[20],
                 "criterion": ['absolute_error']}
    },
    "GradientBoostingRegressor" : {
        "model": GradientBoostingRegressor(),
        "params":{'loss':['squared_error', 'absolute_error', 'huber', 'quantile'],
                 'n_estimators':[10],
                  'criterion':['friedman_mse', 'squared_error']
                 }
    }
   
                   } 
```

```python
best_estimators = {}
scores = []

for model_name,model_params in model_selection.items():
    print(f'performing GridSearchCV {model_name}...')
    clf = GridSearchCV(model_params['model'],model_params['params'],cv=5,scoring='neg_mean_squared_error')
    clf.fit(X,y)
    best_estimators[model_name] = clf.best_estimator_
    print(f'Beat Parameters for {model_name} :  {clf.best_params_}')
    print(f"Best MSE for {model_name}: {clf.best_score_}")
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
final_score = pd.DataFrame(scores,columns=['model','best_score','best_params'])
final_score
```

* Here is built in Ensemble technique to get the best score and parameters...
* So, we can see L2 Regularization or Ridge regression is the best model and the hyperparameter is:

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
model = Ridge(alpha= 7)
model.fit(x_train,y_train)
pre = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,pre)
score
```
